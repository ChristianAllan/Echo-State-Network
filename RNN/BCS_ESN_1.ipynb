SINAL APRBS DO BCS:

```python
# Parâmetros do sinal
t_final = 500                       # duração total
dt = 1/60                           # passo de tempo
min_value, max_value = 35, 60       # limites
min_step, max_step = 100/60, 700/60 # tempo entre trocas

# Vetor de tempo
t = np.arange(0, t_final, dt)

# Inicialização do sinal
freq = np.zeros_like(t)

# Geração do sinal APRBS
current_value = np.random.uniform(min_value, max_value)
time_next_change = 0

for i in range(len(t)):
    if t[i] >= time_next_change:
        current_value = np.random.uniform(min_value, max_value)
        time_next_change = t[i] + np.random.uniform(min_step, max_step)
        #print(time_next_change-t[i])
    freq[i] = current_value
```

```python
# Plot do sinal
plt.figure(figsize=(10, 5))
plt.plot(t, freq, label="Freq APRBS", color='b')
plt.xlabel("Tempo")
plt.ylabel("FREQUÊNCIA")
plt.title("Sinal de frequência APRBS (35 a 60 Hz)")
plt.grid(True)
plt.legend()
plt.show()
```
